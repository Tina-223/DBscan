```python
import pandas as pd
import numpy as np
from gensim.models import Word2Vec

df = pd.read_csv('all_drinks.csv')

df = df.drop(['strCategory','dateModified','strDrinkThumb', 'strGlass', 'strIBA', 'strInstructions', 'strMeasure1', 'strMeasure2', 'strMeasure3', 'strMeasure4', 'strMeasure5', 'strMeasure6', 'strMeasure7', 'strMeasure8',
                          'strMeasure9', 'strMeasure10', 'strMeasure11', 'strMeasure12', 'strMeasure13', 'strMeasure14', 'strMeasure15', 'strVideo', 'strIngredient13', 'strIngredient14', 'strIngredient15'], axis=1)
df.head()
```

```python
alcoholic = {'Alcoholic' : 2, 'Optional alcohol': 1, 'Non alcoholic': 0}
df['strAlcoholic'] = df['strAlcoholic'].map(alcoholic)

df3 = pd.DataFrame(columns = ['final'])

final = []
for row in range(df.shape[0]):
  ingredients = []
  for i in range(1, 13):
    column = 'strIngredient'+ str(i)
    ingr = df.loc[row][column]
    if str(ingr) == '' or str(ingr) == 'nan':
      continue
    elif ingr not in ingredients :
      ingredient = str(ingr).lower()
      ingredients.append(ingredient)
  final.append(ingredients)
  
model = Word2Vec(sentences=final, size=100, window=5, min_count=5, workers=4, sg=0)
```

```python
model_result = model.wv.most_similar('orange')
print(len(model_result))
print(model_result)
```

```python
from sklearn.cluster import DBSCAN

dff = pd.DataFrame(model_result)
df_train = dff.drop(dff.columns[[0]], axis=1)

clustering = DBSCAN(eps= 0.005, min_samples=3).fit(df_train)
df_train.loc[:,'Cluster'] = clustering.labels_
df_train.Cluster.value_counts().to_frame()
```
